```python
year = []
month= []
day  = []
dates= []

for i, row in df.iterrows():
    try:
        temp = datetime.strptime(row.dtDeclaracaoDispensa,'%Y-%m-%dT%H:%M:%S')
        year.append(temp.year)
        month.append(temp.month)
        day.append(temp.day)
        dates.append(temp)
        
    except Exception as e:
        print(e)
        print(temp)
        print(row)
        break
        
    if i%500000 == 0:
        print(i)
```

```python
df['year'] = year
df['month']=month
df['day']  = day
df['dates']=dates
```
